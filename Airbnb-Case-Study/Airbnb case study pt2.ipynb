```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
airbnb = pd.read_csv('airbnb prices.csv')
```

```python
#Plot a histogram of reviews where bins are 20 and range is between 0 to 100
plt.style.use('fivethirtyeight')
plt.hist(airbnb.reviews, bins = 20, range = (0, 100), rwidth=0.95, color='#607c8e')
plt.ylabel('Number of reviews')
plt.xlabel('frequency of reviews')
plt.title("Reviews's Chart")
plt.grid(alpha=0.75)
plt.show()
```

```python
#import the relevant sklearn libraries frome splitting the data test and train as well as for linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import seaborn as seabornInstance 
```

```python
#create the X parameter (indepent variable) by taking the column reviews, overall_satisfaction, accommodates and bedroom
"""
Now we will select the attributes.It is a an independent varaibles which is X is most cases while label
is the dependent variable which is Y in most cases.
"""
# X is an attribute (independent)
X = airbnb[['reviews', 'overall_satisfaction', 'accommodates', 'bedrooms']]
# Y is an label (dependent)
#Create a Y parameter by taking the price
y = airbnb['price']
```

```python
#split the data in train and test by holding tghe 30 % dataset for testing and makes this as a reproducible output
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
#initalize and fit the linear regression.
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns =['Coefficent'])
print(regressor.intercept_)
print(regressor.coef_)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient']) 
```

```python
coeff_df
```

```python
#predicting
y_pred = regressor.predict(X_test)
```

```python
#create a dataframe to check the actual and predicted price
df = pd.DataFrame({"Actual": y_test, "Predict": y_pred})
```

```python
df.head(12)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
